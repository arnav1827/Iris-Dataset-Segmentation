```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
data = pd.read_csv("iris_dataset.csv")
data.head()
```

```python
x = data.iloc[:,:2]
y = data.iloc[:,2:4]
x,y
```

```python
plt.scatter(x['sepal_length'], x['sepal_width'])
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
```

```python
plt.scatter(y['petal_length'], y['petal_width'])
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
```

```python
# Clustering for Sepal width and length 
kmeans = KMeans(2)
kmeans.fit(x)
```

```python
clustered_data_x = x.copy()
clustered_data_x['Cluster_pred'] = kmeans.fit_predict(x)
clustered_data_x
```

```python
plt.scatter(x['sepal_length'], x['sepal_width'], c=clustered_data_x['Cluster_pred'], cmap='rainbow')
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.title("Sepal")
```

```python
# Standardizing the cluster grp to reduce redundancy as observed from the above graph around Sepal-length 6
from sklearn import preprocessing
x_scaled = preprocessing.scale(x)
x_scaled
```

```python
kmeans_scaled = KMeans(2)
kmeans_scaled.fit(x_scaled)
```

```python
clustered_data_x_scaled = x.copy()
clustered_data_x_scaled['Cluster_pred'] = kmeans_scaled.fit_predict(x_scaled)
clustered_data_x_scaled
```

```python
plt.scatter(clustered_data_x_scaled['sepal_length'], clustered_data_x_scaled['sepal_width'], c=clustered_data_x_scaled['Cluster_pred'], cmap='rainbow')
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.title("Sepal")
```

```python
# Clustering for Petal Length and width
kmeansy = KMeans(2)
kmeansy.fit(y)
```

```python
clustered_data_y = y.copy()
clustered_data_y['Cluster_pred'] = kmeansy.fit_predict(y)
clustered_data_y
```

```python
plt.scatter(clustered_data_y['petal_length'], clustered_data_y['petal_width'], c=clustered_data_y['Cluster_pred'], cmap='rainbow')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.title("Petal")
```

```python
# Standarding to try to improve the cluster grp
y_scaled = preprocessing.scale(y)
y_scaled
```

```python
kmeansy_scaled = KMeans(2)
kmeansy_scaled.fit(y_scaled)
```

```python
clustered_data_y_scaled = y.copy()
clustered_data_y_scaled['Cluster_pred'] = kmeansy_scaled.fit_predict(y_scaled)
clustered_data_y_scaled
```

```python
plt.scatter(clustered_data_y_scaled['petal_length'], clustered_data_y_scaled['petal_width'], c=clustered_data_y_scaled['Cluster_pred'], cmap='rainbow')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.title("Petal")
```

```python
# Improving Clustering using Elbow Method
# solving for Sepal
wscc = []
for i in range(1,15):
    kmeans = KMeans(i)
    kmeans.fit(x_scaled)
    wscc_iter = kmeans.inertia_
    wscc.append(wscc_iter)
wscc
```

```python
number_clusters = range(1,15)
plt.plot(number_clusters, wscc)
plt.xlabel("No of Clusters")
plt.ylabel("WSCC")
```

```python
# Applying for y
wsccy = []
for i in range(1,15):
    kmeans = KMeans(i)
    kmeans.fit(y_scaled)
    wscc_iter_y = kmeans.inertia_
    wsccy.append(wscc_iter_y)
wsccy
```

```python
number_clusters = range(1,15)
plt.plot(number_clusters, wsccy)
plt.xlabel("No of Clusters")
plt.ylabel("WSCC")
```

```python
# It can be seen from the above Elbow Graph that 3-5 clusters would be great for Sepal whereas 3 cluster grps would be for Petal
# Therefor trying both for 3 cluster grps.
kmeans_x_3 = KMeans(3)
kmeans_x_3.fit(x_scaled)
clustered_data_x_3 = x.copy()
clustered_data_x_3['Cluster_pred'] = kmeans_x_3.fit_predict(x_scaled)
plt.scatter(clustered_data_x_3['sepal_length'], clustered_data_x_3['sepal_width'], c=clustered_data_x_3['Cluster_pred'], cmap='rainbow')
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.title("Sepal")
```

```python
kmeans_y_3 = KMeans(3)
kmeans_y_3.fit(y_scaled)
clustered_data_y_3 = y.copy()
clustered_data_y_3['Cluster_pred'] = kmeans_y_3.fit_predict(y_scaled)
clustered_data_y_3
plt.scatter(clustered_data_y_3['petal_length'], clustered_data_y_3['petal_width'], c=clustered_data_y_3['Cluster_pred'], cmap='rainbow')
plt.xlabel("Petal Length")
plt.ylabel("Petal Width")
plt.title("Petal")
```
